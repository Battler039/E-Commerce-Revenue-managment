## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
df=ECom_Data
df
#used to rename the dataset as df
```

```python
df['Revenue']=((df.discounted_price>600).astype(int)*df.discounted_price*25/100 +\
               ((df.discounted_price>350) | (df.discounted_price)<=600).astype(int)*df.discounted_price*15/100 +\
               ((df.discounted_price>100) | (df.discounted_price)==350).astype(int)*df.discounted_price*10/100 +\
               ((df.discounted_price>50).astype(int)*df.discounted_price*5/100))
df['Revenue']
#create Revenue column,and apply each conditions
# astype(int) used to convert one datatype into another
df #print the dataset with the column Revenue
```
